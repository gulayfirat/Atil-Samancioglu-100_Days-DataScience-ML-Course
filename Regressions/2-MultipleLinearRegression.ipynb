## CREATE REGRESSION

### 1.Step: independent and dependent features

```python
# if df["Exam Score"] in last index
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
#X=df[["Study Hours","Sleep Hours","Attendance Rate","Social Media Hours"]]
#y=df["Exam Score"]
```

```python
X.head()
```

```python
y.head()
```

### 2.Step: Test - Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.25, random_state=15)
```

### 3.Step: StandartScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### 4.Step: Create Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

### 5.Step: Predicts

```python
X_test[0]
```

```python
df.iloc[0]
```

```python
new_student=[[5,7,90,4]]
```

```python
new_student_scaled=scaler.transform(new_student)
```

```python
regression.predict(new_student_scaled)
```

### 6.Step: Prediction & Regression Success Metrics

```python
y_pred=regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse=mean_squared_error(y_test, y_pred)
mae=mean_absolute_error(y_test, y_pred)
print("MSE: ", mse)
print("MAE: ", mae)
```

### 7.Step: R2 Score

```python
score=r2_score(y_test, y_pred)
print("r2 score: ",score)
```

### 8.Step: Adjusted R2 Score

```python
# 1 - (1-model.score(X, y))*(len(y)-1)/(len(y)-X.shape[1]-1)
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
