```python
data2= data1**2
data2
```

```python
plt.subplots()
```

```python
my_fig, my_axes= plt.subplots()
my_axes.plot(data1, data2, color="#1E0703", alpha=0.5)
my_axes.plot(data2, data1,color="#53CA0A")
plt.show()
```

```python
new_fig,new_axes= plt.subplots()
new_axes.plot(data1, data1+4, color="blue", linewidth=1.0)
new_axes.plot(data1, data1+5, color="yellow", linewidth=4.0)
new_axes.plot(data1, data1+6, color="#c85d09", linestyle="-.")
new_axes.plot(data1, data1+7, color="#000000", linestyle="--", marker="o", markersize=8, markerfacecolor="red")
new_axes.plot(data1, data1+8, color="#000000", linestyle="--", marker="+", markersize=10, markerfacecolor="red")


plt.show()
```
