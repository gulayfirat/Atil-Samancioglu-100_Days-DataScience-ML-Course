```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
age_list=[10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list=[20,60,80,85,86,87,70,90,95,90]
```

## plot - show - savefig

```python
plt.plot(age_list, weight_list, "y") # y: yellow, g: green, b:blue, r: red ....
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show() 
# plt.savefig("test.png") # save to local
```

## working with numpy

```python
np_age_list=np.array(age_list)
np_age_list
```

```python
np_weight_list=np.array(weight_list)
np_weight_list
```

```python
plt.plot(np_age_list, np_weight_list, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

## linspace

```python
numpy_arr1=np.linspace(1,10,20)  # It generates 20 numbers from 1 to 10, and the numbers increase evenly.
numpy_arr1
```

```python
numpy_arr2= numpy_arr1**3
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b*")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b--")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b.-")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b-")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b*-")
plt.show()
```

```python
plt.plot(numpy_arr2, numpy_arr1, "b*-")
plt.show()
```

## subplot

```python
plt.subplot(1,2,1) # 1: 1 row, 2: 2 column, 1: first graph
plt.plot(numpy_arr1, numpy_arr2, "r*-")
plt.subplot(1,2,2) # 1: 1 row, 2: 2 column, 1: second graph
plt.plot(numpy_arr2, numpy_arr1, "g--")
plt.show()
```

## figure

```python
my_figure=plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.3,0.3]) # 0.3:width , 0.3:height
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```
