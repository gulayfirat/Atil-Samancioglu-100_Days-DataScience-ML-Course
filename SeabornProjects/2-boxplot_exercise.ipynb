```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Box Plot

#### median, quartile (25%, 50%, 75%), min, max=1.5 IQR, outlier

```python
data=np.array([5,7,9, 15, 20,22,25, 30,32,35,37,40,50,55,60, 100])
plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Blot")
plt.ylabel("Data Value")
#plt.grid(True)
plt.show()

#Center line: Median
#Blue Frame: Data density
#Blue Frame Top Line: %75
#Blue Frame Bottom Line: %25
#Top line: Max data
#Bottom Line: Min data
#Circle: Outlier(100)
```
