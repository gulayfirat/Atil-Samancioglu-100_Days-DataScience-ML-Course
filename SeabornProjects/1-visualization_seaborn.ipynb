# Seaborn

## set_style

```python
sns.set_style("white") # themes: whitegrid, darkgrid, white,white 
sns.scatterplot(x="Height", y="Weight", data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
sns.set_style("darkgrid") # themes: whitegrid, darkgrid, white,white 
sns.scatterplot(x="Height", y="Weight", data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
data.info()
```

## hue

```python
sns.set_style("whitegrid") # themes: whitegrid, darkgrid, white,white 
sns.scatterplot(x="Height", y="Weight", data=data, hue="Sex")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
data.columns
```

## unique

```python
data["Medal"].unique()
```

```python
data["Sex"].unique()
```

```python
sns.set_style("whitegrid") # themes: whitegrid, darkgrid, white,white 
sns.scatterplot(x="Height", y="Weight", data=data, hue="Sex", style="Medal", size="Age")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") # themes: whitegrid, darkgrid, white,white 
sns.lineplot(x="Height", y="Weight", data=data, hue="Sex")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

## displot

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex", data=data)
plt.ylabel("Frequency")
plt.title("Athletes Height Distrubition")
plt.show()
```

## kind="kde"

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex", data=data, kind="kde") # kde: kernel denstiy estimate
plt.ylabel("Frequency")
plt.title("Athletes Height Distrubition")
plt.show()
```

## barplot

```python
sns.barplot(x="Medal",y="Height", hue="Sex", data=data)
plt.title("Medals by Height")
plt.show()
```

## catplot (cat: categority)

```python
data.columns
```

```python
data["Season"].unique()
```

```python
sns.catplot(x="Medal",y="Height", hue="Sex", col="Season" ,data=data)
plt.show()
```

### kind: bar

```python
sns.catplot(x="Medal",y="Height", hue="Sex", col="Season" ,data=data, kind="bar")
plt.show()
```
