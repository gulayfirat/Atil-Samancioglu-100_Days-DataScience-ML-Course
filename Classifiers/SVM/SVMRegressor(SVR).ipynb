```python
X= df.drop("price", axis=1)
y=df["price"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
for col in ["cut", "color", "clarity"]:
    X_train[col]=label_encoder.fit_transform(X_train[col]) # for X_train -> used fit_transform
    X_test[col]=label_encoder.transform(X_test[col]) # for X_test -> used transform
```

```python
X_train.head()
```

```python
X_train["cut"].value_counts()
```

```python
X_test["cut"].value_counts()
```

```python
X_train.info()
```

```python
X_train.describe()
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
#Create Module
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,  mean_absolute_error, r2_score
```

```python
linear= LinearRegression()
linear.fit(X_train_scaled, y_train)
y_pred= linear.predict(X_test_scaled)
mae= mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
score= r2_score(y_test, y_pred)
print("Mean absolute error: ", mae)
print("Mean squared error: ", mse)
print("R2 score", score)
plt.scatter(y_test, y_pred)
```

```python
# SVR (Support Vector Regressor)
from sklearn.svm import SVR
```

```python
svr= SVR()
svr.fit(X_train_scaled, y_train)
y_pred= svr.predict(X_test_scaled)
mae= mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
score= r2_score(y_test, y_pred)
print("Mean absolute error: ", mae)
print("Mean squared error: ", mse)
print("R2 score", score)
plt.scatter(y_test, y_pred)
```

```python
# hyperparameter tuning for SVR
# GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    "C":[0.1, 1, 10, 100, 1000],
    "gamma":[0.1, 0.01, 0.001],
    "kernel": ["rbf", "linear", "poly"],
}
```

```python
grid= GridSearchCV(estimator=SVR(), ,param_grid= param_grid, n_jobs=-1, verbose= 3)
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
y_pred= grid.predict(X_test_scaled)
mae= mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
score= r2_score(y_test, y_pred)
print("Mean absolute error: ", mae)
print("Mean squared error: ", mse)
print("R2 score", score)
plt.scatter(y_test, y_pred)
```
